```python
%%time
import glob
import os
import pandas as pd
import sj2psi

dirs = {'pe': '/home/obotvinnik/projects/singlecell_pnms/analysis/singlecell_pnms_pe_v6',
        'pe_redo': '/home/obotvinnik/projects/singlecell_pnms/analysis/singlecell_pnms_pe_redo_v1',
        'se': '/home/obotvinnik/projects/singlecell_pnms/analysis/singlecell_pnms_se_v6'}

MIN_UNIQUE_JUNCTION_READS = 10
MIN_MULTIMAP_JUNCTION_READS = 20


# sj_raw = []
dfs = []
for name, DIR in dirs.items():
    for i, filename in enumerate(glob.iglob("{}/*rmRep*SJ.out.tab".format(DIR))):
        df = sj2psi.read_sj_out_tab(filename)
        basename = os.path.basename(filename)
        sample_id = '_'.join(basename.split('_')[:2])
        print i, sample_id
#         min_unique_reads = df.unique_junction_reads >= MIN_UNIQUE_JUNCTION_READS
#         min_multimap_reads = df.multimap_junction_reads >= MIN_MULTIMAP_JUNCTION_READS
    #     print '\t', df.shape
#         df = df.loc[min_unique_reads]
        df['sample_id'] = sample_id
    #     print '\t', df.shape
        df['run_name'] = name
        dfs.append(df)
sj_raw = pd.concat(dfs, ignore_index=True)
```

## Remove the samples in `pe_redo` from `pe`

```python
redo_ids = sj_raw.loc[sj_raw.run_name == 'pe_redo'].sample_id.unique()
to_drop = sj_raw.index[(sj_raw.run_name == 'pe') & (sj_raw.sample_id.isin(redo_ids))]
print sj_raw.shape
sj = sj_raw.drop(to_drop, errors='ignore')
print sj.shape
sj.head()
```

### Remove `pe_redo`, `pe` run names

```python
sj = sj.drop(['run_name'], axis=1, errors='ignore')
sj.head()
```

```python
sj['intron_location'] = sj.chrom.astype(str) + ':' \
    + sj.intron_start.astype(int).astype(str) + '-' \
    + sj.intron_stop.astype(int).astype(str) + ':' \
    + sj.strand.astype(str)

```

### Distribution of number of reads per junction

```python
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
sns.set(style='ticks', context='talk')
%matplotlib inline

reads = sj.unique_junction_reads[sj.unique_junction_reads > 0]
# bins = np.linspace(0, 6, 1000)

fig, ax = plt.subplots(figsize=(4, 3))

sns.distplot(np.log10(reads+1), kde=False, color='darkgrey')

ymin, ymax = ax.get_ylim()
xticks = ax.get_xticks()
yticks = ax.get_yticks()

yticklabels = [int(i/1e5) for i in yticks]

xticklabels = ['$10^{{{}}}$'.format(int(i)) for i in xticks]
ax.vlines(1, 0, ymax, linestyle='dashed')
ax.set(xlabel='Uniquely Mapped Junction Reads', xticklabels=xticklabels, yticklabels=yticklabels,
       ylabel='100,000s of junctions')
#bins=[-1, 0, 10, 100, 1000, 1e4, 1e5, 1e6, 1e7], kde=False)
# ax.set_xlim(0, 50)
sns.despine()
fig.tight_layout()
fig.savefig('/home/obotvinnik/Dropbox/figures2/singlecell_pnm/uniquely_mapped_junction_reads.pdf')
```
