```python
import pandas as pd
```

```python
se_gencode_filename = '/home/obotvinnik/projects/singlecell_pnms/analysis/outrigger_v2/index/se/event.sorted.gencode.v19.bed'
se_gencode = pd.read_table(se_gencode_filename, header=None)
se_gencode.head()
```

```python
def split_gtf_attributes(attributes):
    split = attributes.split('; ')
    pairs = [x.split(' ') for x in split]
    no_quotes = [map(lambda x: x.strip('";'), pair) for pair in pairs]
    mapping = dict(no_quotes)
    return mapping
```

```python
%%time

attributes = se_gencode[14].map(split_gtf_attributes).apply(pd.Series)
print(attributes.shape)
attributes.head()
```

```python
attributes.index = se_gencode[3]
attributes.index.name = 'event_id'
attributes.head()
```

```python
attributes['ensembl_id'] = attributes['gene_id'].str.split('.').str[0]
attributes.head()
```

```python
bad_transcript_types = 'snoRNA', 'snRNA', 'ncRNA_host', 'antisense', 'miRNA', 'IG_V_pseudogene', \
    '3prime_overlapping_ncrna', 'misc_RNA', 'rRNA', 'sense_intronic', 'sense_overlapping'

good_transcript_types = 'lincRNA', 'protein_coding', 'pseudogene', 'processed_transcript'
```

```python
attributes.shape
```

```python
attributes_good_transcripts = attributes.query('transcript_type not in @bad_transcript_types')
print(attributes_good_transcripts.shape)

attributes_good_transcripts.head()
```

```python
attributes_grouped = attributes_good_transcripts.groupby(level=0, axis=0).apply(
    lambda df: df.apply(lambda x: ','.join(map(str, set(x.values)))))

print(attributes_grouped.shape)
attributes_grouped.head()
```

```python
from outrigger.region import Region
```

```python
%%time

se_exon_bed_template = '/home/obotvinnik/projects/singlecell_pnms/analysis/outrigger_v2/index/se/exon{}.bed'

exons = {}

for i in range(1, 4):
    exon = 'exon{}'.format(i)
    print(exon)
    bed = se_exon_bed_template.format(i)
    exon_df = pd.read_table(bed, names=['chrom', 'start', 'stop', 'name', 'score', 'strand'], index_col='name')
    exon_regions = exon_df.apply(lambda row: Region('exon:{chrom}:{start}-{stop}:{strand}'.format(
                chrom=row.chrom, start=row.start+1, stop=row.stop, strand=row.strand)), axis=1)
    exons[exon] = exon_regions
exon_regions_df = pd.DataFrame(exons)
print(exon_regions_df.shape)
```
