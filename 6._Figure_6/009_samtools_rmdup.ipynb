```python
import glob
import os
import qtools
```

```python
bams = glob.glob('*bam')

template = 'samtools rmdup {folder}/{bam} {folder}/rmdup/{bam}'

commands = []
for bam in bams:
    command = template.format(folder=folder, bam=bam)
    commands.append(command)
```

```python
project = 'singlecell_pnms'
```

```python
cd ~/projects/$project/scripts
```

```python
jobname = '{project}_samtools_rmdup'.format(project=project)

qtools.Submitter(commands, jobname, array=True, walltime='2:00:00')
```

## Resubmit the .bam files that didn't end successfully.

```python
tails = ! tail -n 1 $jobname*err* 
```

```python

def get_uncompleted_jobs(lines):
    for line in lines:
        if line.startswith('=='):
            index = int(line.split()[1].split('-')[-1])
        elif line.startswith('['):
            if 'in library' in line:
                continue
            else:
                yield index
uncompleted_jobs = tuple(sorted(get_uncompleted_jobs(tails)))
uncompleted_jobs
```

```python
len(commands)
```

```python
unfinished_commands = [command for i, command in enumerate(commands) if (i+1) in uncompleted_jobs]
unfinished_commands
```

Oh right these are single-ended and you have to do `samtools rmdup` differently...

```python
single_end_commands = [x.replace('rmdup', 'rmdup -s') for x in unfinished_commands]
single_end_commands[:5]
```

```python
jobname = '{project}_samtools_rmdup_se'.format(project=project)

qtools.Submitter(single_end_commands, jobname, array=True, walltime='1:00:00')
```
