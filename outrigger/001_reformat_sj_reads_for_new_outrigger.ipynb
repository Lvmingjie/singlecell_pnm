```python
import pandas as pd
```

```python
junction_reads = pd.read_csv('/projects/ps-yeolab/obotvinnik/singlecell_pnms/csvs_for_paper/sj_raw.csv')
print(junction_reads.shape)
junction_reads.head()
```

```python
renamer = dict((x, x.replace('intron', 'junction')) for x in junction_reads if 'intron' in x)
renamer
```

```python
junction_reads_renamed = junction_reads.rename(columns=renamer)
print(junction_reads_renamed.shape)
junction_reads_renamed.head()
```

```python
junction_reads_renamed_id = junction_reads_renamed.copy()
junction_reads_renamed_id['junction_id'] = 'junction:'  + junction_reads_renamed_id['junction_location']
junction_reads_renamed_id.head()
```

```python
junction_reads_renamed_id['exon_start'] = junction_reads_renamed_id['junction_stop'] + 1
junction_reads_renamed_id['exon_stop'] = junction_reads_renamed_id['junction_start'] - 1
junction_reads_renamed_id.head()
```

```python
junction_reads_renamed_id['reads'] = junction_reads_renamed_id['unique_junction_reads'] + junction_reads_renamed_id['multimap_junction_reads']
junction_reads_renamed_id.head()
```

```python

junction_reads_renamed_id.to_csv('/projects/ps-yeolab/obotvinnik/singlecell_pnms/csvs_for_paper/junction_reads_use_multimapping.csv', index=False)
```
